```
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

# Carregar o dataset
url = "https://raw.githubusercontent.com/ybifoundation/Dataset/main/Salary%20Data.csv"
df = pd.read_csv(url)

# Separar variáveis (corrigido para 'Experience Years')
X = df[['Experience Years']]
y = df['Salary']

# Criar e treinar o modelo
model = LinearRegression()
model.fit(X, y)

# Previsões
y_pred = model.predict(X)

# Métrica R²
print(f"\nCoeficiente de determinação (R²): {r2_score(y, y_pred):.4f}")

# Gráfico
plt.figure(figsize=(10,6))
plt.scatter(X, y, color='blue', label='Dados reais')
plt.plot(X, y_pred, color='red', label='Regressão Linear')
plt.xlabel('Anos de Experiência')
plt.ylabel('Salário')
plt.title('Regressão Linear: Experiência vs Salário')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
